Cleaning and Normalizing Text Data

```python
for index,text in enumerate(df['reviews'][35:150]):
  print('Review %d:\n'%(index+1),text)
```

```python
df['cleaned']=df['reviews'].apply(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '', x))
df
```

```python
df['cleaned']=df['cleaned'].apply(lambda x: re.sub('\w*\d\w*','', x))
df
```

```python
df['cleaned'] = df['cleaned'].str.replace('[^\w\s]','')  # Removing other punctuations
df
```

```python
df 
```

```python
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')

model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
def sentiment_score(cleaned):
    tokens = tokenizer.encode(cleaned, return_tensors='pt')
    result = model(tokens)
    return int(torch.argmax(result.logits))+1
```

```python
sentiment_score(df['cleaned'].iloc[1])
```

```python
df['sentiment_score'] = df['cleaned'].apply(lambda x: sentiment_score(x[:512]))
```
